# Data Import Code Snippet
Data to be imported in .xls format from FRED.
<br>
Stores data as exchange rate (float) and date (date_time).






```python
from xlrd import open_workbook
import datetime

class ExR(object):
    def __init__(self, dt, rate):
        self.dt = dt
        self.rate = rate


    def __str__(self):
        return("ExR object:\n"
               "  Observation_date = {0}\n"
               "  DEXUSEU = {1}\n"
               .format(self.dt, self.rate))
    
    
def convert_excel_time(excel_time):
    temp = datetime.date(1899, 12, 30)
    delta = datetime.timedelta(days=excel_time)
    return temp+delta

wb = open_workbook('DEXUSEU.xls')
for sheet in wb.sheets():
    number_of_rows = sheet.nrows
    number_of_columns = sheet.ncols

    items = []

    rows = []
    for row in range(11, number_of_rows):
        values = []
        value  = (sheet.cell(row,0).value)
        value = int(value)
        value = convert_excel_time(value)
        values.append(value)
        value  = (sheet.cell(row,1).value)
        value = float(value)
        if value==0:
            continue
        values.append(value)
        item = ExR(*values)
        items.append(item)

for item in items:
    print (item)
    print
```

```python
from matplotlib import pyplot as plt
from matplotlib.dates import date2num

y = []
xdate = []
x = []

for item in items:
    d=date2num(item.dt)
    x.append(d)
    y.append(item.rate)

fig = plt.figure(figsize=(20, 14))

graph = fig.add_subplot(111)

# Plot the data as a red line with round markers
graph.plot(x,y,'r-x')


plt.show()
```
